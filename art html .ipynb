```python
#importing the library files for the design
import  pygame
import pygame.surfarray as surfarray
import matplotlib.pyplot as plt
import random


#initialization of the pygame
pygame.init()

#size of the generated arts screen
width=1000
height=1000
screen=pygame.display.set_mode((width,height))

#set the number of circles and squares
number_squares=900
number_circles=500

#set the numbers of squares of the top left corner
squares=[]
for i in range(number_squares):
    x=random.randint(0,width/2)
    y=random.randint(height/2,height)
    size=random.randint(10,20)
    color=(random.randint(0,255), random.randint(0,255),random.randint(0,255))
    squares.append((x,y,size,color))
#geneate the number of squares of the bottom right corner

for i in range(number_squares):
    x=random.randint(width/2,width)
    y=random.randint(0,height/2)
    size=random.randint(10,40)
    color=(random.randint(0,255),random.randint(0,255),random.randint(0,255))
    squares.append((x,y,size,color))
#draw the squares
screen.fill((255,255,255))
for square in squares:
    pygame.draw.rect(screen,square[3],(square[0],square[1],square[2],square[2]))

#generate the circles of the bottom right corner
circles=[]
for i in range(number_circles):
            x=random.randint(width//2,width)
            y=random.randint(height//2,height)
            radius=random.randint(4,10)
            color=(random.randint(0,255),random.randint(0,55),random.randint(0,255))
            circles.append((x,y,radius,color))
            
#generate the circles for top left corner
for i in range(number_circles):
            x=random.randint(0,width/2)
            y=random.randint(0,height/2)
            radius=random.randint(4,30)
            color=(random.randint(0,230),random.randint(0,35),random.randint(0,20))
            circles.append((x,y,radius,color))
#draw the circles
for circle in circles:
    pygame.draw.circle (screen,circle[3],(circle[0],circle[1]),circle[2])
                        
#convert the pygame surface to a Numpy array
pygame.display.flip();
pygame.surfarray.use_arraytype("numpy")
image=surfarray.array3d(screen)

#display in image using matplotlib
plt.imshow(image)
plt.axis("off")
plt.show()

#quit pygame
pygame.quit()
```
